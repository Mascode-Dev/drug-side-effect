```python
# Avoir la liste des side_effect uniques
side_effects = df['side_effect']
side_effects = side_effects.unique()
side_list = []
for s in side_effects:
    tmp_list = s.split(';')
    for i in range(len(tmp_list)):
        if tmp_list[i] not in side_list:
            side_list.append(tmp_list[i])
        else:
            continue
        
print(side_list)
print(f"Nombre de side_effects uniques: {len(side_list)}")
```

Since we want to do some textual recognition model, we will drop some useless column like stitch_id or description (not exploitable)

```python
df[['drug_name', 'name']].head()
# Since we have very similar names, we will drop the 'name' column
df = df.drop(columns=['name'])
```

```python
# We are dropping all ids columns since they are not useful for our analysis
df = df.drop(columns=['drugbank_id', 'stitch_id'])
```

```python
# Description drop because unexploitable text data
df = df.drop(columns=['description'])
```

```python
# NaN values
print(df.isna().sum())

#Since we are working on text value we need to drop rows with NaN values
df = df.dropna()
```
